```python
# test the problems with exploration
class Node():
    def __init__(self, depth, value):
        self.count = 0
        self.depth = depth
        self.value = value
        self.children = {}
        
    def add_child(self, depth, value):
        self.children[value] = Node(depth, value)
        return self.children[value]
        
    def remove_child(self, node):
        try:
            self.children.remove(node)
        except:
            print("child not found")
    
class Trie():
    def __init__(self):
        self.root = Node(0, "root")
        self.depth = 0
        self.nodes = 1
        
    def add_path(self, path):
        if len(path) != 4:
            return
        traverse = self.root
        depth = 1
        for node in path:
            try:
                traverse = traverse.children[node]
            except:
                self.nodes += 1
                if (depth > self.depth):
                    self.depth += 1
                traverse = traverse.add_child(depth, node)
            depth += 1
        traverse.count += 1
        
    def visualize(self):
        self.print_arr = [0,0,0,0,0]
        self.update_lists(self.root)
        return self.print_arr
                
    def update_lists(self, node):
        for x in list(node.children.keys()):
            self.update_lists(node.children[x])
            try:
                self.print_arr[node.depth + 1] += 1
            except:
                self.print_arr[node.depth + 1] = 1

import random
from operator import add
averages_origin = [0,0,0,0,0]
averages_no_sl_nomask = [0,0,0,0,0]
averages_no_sl_mask = [0,0,0,0,0]
averages_sl_rl_mask = [0,0,0,0,0]
averages_sl_rl_nomask = [0,0,0,0,0]
for query in range(len(queries)):
    orig_reward_nodes = Trie()
    for path in orig_reward_queries[list(orig_reward_queries.keys())[query]]:
        orig_reward_nodes.add_path(path.split('\n')[0].split('	'))
    averages_origin = list(map(add, averages_origin, orig_reward_nodes.visualize()))
    
    no_sl_nomask_nodes = Trie()
    for path in no_sl_queries_nomask[list(no_sl_queries_nomask.keys())[query]]:
        no_sl_nomask_nodes.add_path(path.split('\n')[0].split('	'))
    averages_no_sl_nomask = list(map(add, averages_no_sl_nomask, no_sl_nomask_nodes.visualize()))
    
    no_sl_mask_nodes = Trie()
    for path in no_sl_queries_mask[list(no_sl_queries_mask.keys())[query]]:
        no_sl_mask_nodes.add_path(path.split('\n')[0].split('	'))
    averages_no_sl_mask = list(map(add, averages_no_sl_mask, no_sl_mask_nodes.visualize()))
    
    sl_rl_mask_nodes = Trie()
    for path in sl_rl_queries_mask[list(sl_rl_queries_mask.keys())[query]]:
        sl_rl_mask_nodes.add_path(path.split('\n')[0].split('	'))
    averages_sl_rl_mask = list(map(add, averages_sl_rl_mask, sl_rl_mask_nodes.visualize()))
    
    sl_rl_nomask_nodes = Trie()
    for path in sl_rl_queries_nomask[list(sl_rl_queries_nomask.keys())[query]]:
        sl_rl_nomask_nodes.add_path(path.split('\n')[0].split('	'))
    averages_sl_rl_nomask = list(map(add, averages_sl_rl_nomask, sl_rl_nomask_nodes.visualize()))
    
averages_origin = [x/len(queries) for x in averages_origin]
averages_no_sl_nomask = [x/len(queries) for x in averages_no_sl_nomask]
averages_no_sl_mask = [x/len(queries) for x in averages_no_sl_mask]
averages_sl_rl_mask = [x/len(queries) for x in averages_sl_rl_mask]
averages_sl_rl_nomask = [x/len(queries) for x in averages_sl_rl_nomask]
print("average origin: ")
print(averages_origin)
print("average no sl nomask: ")
print(averages_no_sl_nomask)
print("average no sl mask: ")
print(averages_no_sl_mask)
print("average sl rl nomask: ")
print(averages_sl_rl_nomask)
print("average sl rl mask: ")
print(averages_sl_rl_mask)
```

```python
ds = []
with open("datasets/data_preprocessed/FB15K-237/graph.txt") as input_file:
    csv_file = csv.reader(input_file, delimiter = '\t' )
    for line in csv_file:
        ds.append(line)

connections = {}
for line in ds:
    e1 = line[0]
    r = line[1]
    e2 = line[2]
    try:
        connections[e1] += 1
    except:
        connections[e1] = 1

nodes = {}
for node in list(connections.keys()):
    nodes[node] = [connections[node], min(connections[node], 500)]
```
