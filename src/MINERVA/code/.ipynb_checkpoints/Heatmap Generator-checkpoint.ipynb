```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
data = []
x=[0,2,3,4,5]
for i in x:
    ckpt = []
    with open("/mnt/c/Users/owenb/OneDrive/Documents/GitHub/MINERVA_tf2/output/fb15k-237/ca0f_conv_FB15K-237_paper/checkpoint_sl_epoch_"+str(i)+"/scores.txt","r") as scores:
        chunk = scores.read().split("Score for iteration 2000")[1]
        auc = chunk.split("auc:  ")[2]
        chunk = chunk.split("auc:  ")[1]
        ckpt.append(float(chunk.split("Hits@1:  ")[1].split("\nHits@3")[0])-0.22)
        ckpt.append(float(chunk.split("\nHits@3:  ")[1].split("\nHits@5")[0])-0.33)
        ckpt.append(float(chunk.split("\nHits@5:  ")[1].split("\nHits@10")[0])-0.38)
        ckpt.append(float(chunk.split("\nHits@10:  ")[1].split("\nHits@20")[0])-0.45)
        ckpt.append(float(chunk.split("\nHits@20:  ")[1])-0.5)
        ckpt.append(float(auc)-0.29)
    data.append(ckpt)
data = np.array(data)
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(data, annot=True, linewidths=.5, ax=ax, xticklabels=['Hits@1','Hits@3','Hits@5','Hits@10','Hits@20','auc'], yticklabels=x)#,'no mask','20% mask','40% mask'
```
